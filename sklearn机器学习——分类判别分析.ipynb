```python
# 逻辑回归
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model as lm
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
# 导入数据

df = pd.read_csv("iris.csv")

df.head()
```

```python
# 设置自变量X和因变量Y
X = df.iloc[:,:4]
y = df['species']
```

```python
# 逻辑回归
LogReg = lm.LogisticRegression() # 构建模型对象
LogReg.fit(X,y)   # 传入数据，计算模型
print('拟合得分：',LogReg.score(X,y))  # 拟合得分
print('系数估计：\n',LogReg.coef_)  # 参数估计
print('截距估计：',LogReg.intercept_)  # 参数估计
df['LogReg'] = LogReg.predict(X) # 保存预测结果
x = [[4.8,3.2,1.6,0.5]]
print('样本预测：',x,'--->',LogReg.predict(x))  # 预测测试数据

print('\n预测结果统计（列联表）：\n',pd.crosstab(df['species'],df['LogReg']))
```

```python
# 朴素贝叶斯分类
from sklearn.naive_bayes import GaussianNB

Gnb = GaussianNB() # 构建模型对象
Gnb.fit(X,y)   # 传入数据，计算模型
print('拟合得分：',Gnb.score(X,y))  # 拟合得分
df['Gnb'] = Gnb.predict(X) # 保存预测结果
x = [[4.8,3.2,1.6,0.5]]
print('样本预测：',x,'--->',Gnb.predict(x))  # 预测测试数据
print('\n预测结果统计（列联表）：\n',pd.crosstab(df['species'],df['Gnb']))
```

```python
# 决策树
from sklearn import tree
dTree = tree.DecisionTreeClassifier()  # 决策树分类器对象
dTree.fit(X,y)  # 传入数据，计算模型
print('拟合得分：',dTree.score(X,y))  # 拟合得分
df['dTree'] = dTree.predict(X) # 保存预测结果
x = [[4.8,3.2,1.6,0.5]]
print('样本预测：',x,'--->',dTree.predict(x))  # 预测测试数据
print('\n预测结果统计（列联表）：\n',pd.crosstab(df['species'],df['dTree']))
```

```python
# 随机深林
from sklearn.ensemble import RandomForestClassifier
rndForest = RandomForestClassifier()  # 随机森林分类器对象
rndForest.fit(X,y)  # 传入数据，计算模型

print('拟合得分：',rndForest.score(X,y))  # 拟合得分
df['rndForest'] = rndForest.predict(X) # 保存预测结果
x = [[4.8,3.2,1.6,0.5]]
print('样本预测：',x,'--->',rndForest.predict(x))  # 预测测试数据
print('\n预测结果统计（列联表）：\n',pd.crosstab(df['species'],df['rndForest']))
```

```python
# 支持向量机
from sklearn import svm
svc = svm.SVC()  # 创建分类器对象
svc.fit(X,y)
print('拟合得分：',svc.score(X,y))  # 拟合得分
df['svc'] = svc.predict(X) # 保存预测结果
x = [[4.8,3.2,1.6,0.5]]
print('样本预测：',x,'--->',svc.predict(x))  # 预测测试数据
print('\n预测结果统计（列联表）：\n',pd.crosstab(df['species'],df['svc']))
```

```python
# KNN算法——K近邻
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X,y)

print('拟合得分：',knn.score(X,y))  # 拟合得分
df['knn'] = knn.predict(X) # 保存预测结果
x = [[4.8,3.2,1.6,0.5]]
print('样本预测：',x,'--->',knn.predict(x))  # 预测测试数据
print('\n预测结果统计（列联表）：\n',pd.crosstab(df['species'],df['knn']))
```
